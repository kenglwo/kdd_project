## Run experiments concurrently

```python
from joblib import Parallel, delayed
from tqdm.notebook import tqdm

load_from_json = False
# load_from_json = True
n_rounds = 5
n_jobs = 24
def parallel (f, params_list, params_dict={}, total=None, tqdm_postfix='', leave=True, n_jobs=-1):
    results = []
    with Parallel(n_jobs=n_jobs, require='sharedmem') as p:
        results = p(delayed(f)(*params, **params_dict)
                    for params in tqdm(params_list, postfix=tqdm_postfix, leave=leave, total=total))
    return results
```

```python
def minirocket_train_test(dataset_name, trial_id):
    t_start = time.time()
    X_train, y_train = sktime.datasets.load_UCR_UEA_dataset(dataset_name, split="train", return_X_y=True) 
    
    minirocket = MiniRocket()
    minirocket.fit(X_train)

    t_start_transform_train = time.time()
    X_train_transform = minirocket.transform(X_train)
    time_transform_train = time.time() - t_start_transform_train
    
    classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10), normalize=True)
    classifier.fit(X_train_transform, y_train)
    
    time_train = time.time() - t_start
    
    X_test, y_test = sktime.datasets.load_UCR_UEA_dataset(dataset_name, split="test", return_X_y=True)

    t_start_transform_test = time.time()
    X_test_transform = minirocket.transform(X_test)
    time_transform_test = time.time() - t_start_transform_test
    
    classifier_score = classifier.score(X_test_transform, y_test)
    
    time_total = time.time() - t_start
    
    return {
        'trial_id': trial_id,
        'dataset_name': dataset_name,
        'classifier_score': classifier_score,
        'transformer_name': 'minirocket',
        'time_transform_train': time_transform_train,
        'time_transform_test': time_transform_test,
        'time_train': time_train,
        'time_total': time_total
    }
```

## ROCKET on UCR

```python
def rocket_train_test(dataset_name, trial_id):
    t_start = time.time()
    X_train, y_train = sktime.datasets.load_UCR_UEA_dataset(dataset_name, split="train", return_X_y=True) 
    
    rocket = Rocket()
    rocket.fit(X_train)

    t_start_transform_train = time.time()
    X_train_transform = rocket.transform(X_train)
    time_transform_train = time.time() - t_start_transform_train
    
    classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10), normalize=True)
    classifier.fit(X_train_transform, y_train)
    
    time_train = time.time() - t_start
    
    X_test, y_test = sktime.datasets.load_UCR_UEA_dataset(dataset_name, split="test", return_X_y=True)

    t_start_transform_test = time.time()
    X_test_transform = rocket.transform(X_test)
    time_transform_test = time.time() - t_start_transform_test
    
    classifier_score = classifier.score(X_test_transform, y_test)
    
    time_total = time.time() - t_start
    
    return {
        'trial_id': trial_id,
        'dataset_name': dataset_name,
        'classifier_score': classifier_score,
        'transformer_name': 'rocket',
        'time_transform_train': time_transform_train,
        'time_transform_test': time_transform_test,
        'time_train': time_train,
        'time_total': time_total
    }
```

## Tuning Hyperparameters

```python
from sktime.forecasting.compose import make_reduction
from sktime.forecasting.model_selection import (
    ForecastingGridSearchCV,
    SlidingWindowSplitter,
)
```

```python
start0 = time.time()
best_score = 0
best_params = {'max_dilations_per_kernel': None, 'num_features': None}
Name = ["Beef","Coffee","Car","BirdChicken","Adiac"]
for mc in [16,32]:
    for gm in [1000,10000]:
        for name in Name:
            ava_score = []
            start1 = time.time()
            mir = MiniRocket(max_dilations_per_kernel=mc,num_features=gm)
            X_train, y_train = sktime.datasets.load_UCR_UEA_dataset(name, split="train", return_X_y=True) 
            mir.fit(X_train)
            X_train_transform = mir.transform(X_train)
            classifier.fit(X_train_transform, y_train)
            X_test, y_test = sktime.datasets.load_UCR_UEA_dataset(name, split="test", return_X_y=True)
            X_test_transform = mir.transform(X_test)
            #classifier.score(X_train_transform, y_train)
            ava_score.extend([classifier.score(X_train_transform, y_train)])
        score = mean(ava_score)
        if score > best_score:
            best_score = score
            best_params['max_dilations_per_kernel'] = mc
            best_params['num_features'] = gm
        print(mc,gm,score)
        print(best_params) 
        print(time.time() - start1) 
```
